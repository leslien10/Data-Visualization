```python
import pandas as pd
```

Ex 1

```python
# a.
dt_ap = pd.read_csv('aapl.csv', index_col = [0], parse_dates = [0])
dt_am = pd.read_csv('amzn.csv', index_col = [0], parse_dates = [0])
dt_fb = pd.read_csv('fb.csv', index_col = [0], parse_dates = [0])
dt_go = pd.read_csv('goog.csv', index_col = [0], parse_dates = [0])
dt_nf = pd.read_csv('nflx.csv', index_col = [0], parse_dates = [0])
```

```python
# b.
dt_ap['ticker'] = 'APPL'
dt_am['ticker'] = 'AMZ'
dt_fb['ticker'] =  'FB'
dt_go['ticker'] = 'GOOG'
dt_nf['ticker'] = 'NETF'
```

```python
# c.
faang = pd.concat([dt_ap,dt_am,dt_fb,dt_go,dt_nf], axis = 0, join = 'inner').sort_index()
faang.head()
```

```python
# d.
faang.to_csv('faang.csv')
```

```python
# e.
faang.index = pd.to_datetime(faang.index)
faang['volume'] = faang['volume'].astype(int)
faang.head()
```

```python
faang.sort_values(by=['date', 'ticker'])
faang.head()
```

```python
# f.
faang.sort_values(by='volume', ascending = True).head(7)
```

```python
# a.
dataset = pd.read_csv('covid19_cases.csv')
dataset.head()
```

```python
# b.
dataset['date'] = pd.to_datetime(dataset['dateRep'])
dataset.head()
```

```python
# c.
dataset.set_index('date', inplace = True)
dataset = dataset.sort_index()
dataset.head()
```

```python
# d.
dataset = dataset.replace(['United_States_of_America','United_Kingdom'], ['USA','UK'])
dataset.head()
```

```python
# e.
countries = ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA']
dataset_2 = dataset[dataset['countriesAndTerritories'].isin(countries)]
dataset_2.sort_values('countriesAndTerritories', axis=0, ascending= True).head()
```

```python
# f.
dataset_3 = dataset.pivot(columns='countriesAndTerritories',values='cases').fillna(0)
dataset_3.head()
```
