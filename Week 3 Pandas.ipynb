```python
# Import Dataset
dataset = pd.read_csv('earthquakes.csv')
dataset.head()
```

2. Calculate the mean of mag in Alaska and the distance is smaller than 100km, example:
"62km W of Cantwell, Alaska".

```python
# Hàm lấy số km từ data
def extract(var):
    matches = re.findall(r" - (.*)km", var)
    if not matches:
        return 100
    return int(matches[0])
```

```python
# Xác định km của data
dataset["check_dis"] = [extract(x) for x in dataset["title"]]
```

```python
data_2 = dataset[(dataset["title"].str.contains("Alaska")) & (dataset["check_dis"] < 100)]
data_2.head()
```

```python
# Tính mean
np.mean(data_2["mag"])
```

3. Calculate the mean of mag in each quarter (q1:m1-m3, q2: m4-m6, q3: m7-m9, q4: m10-12), the "time" field stores the milliseconds from 1/1/1970, hint: timestamp.

```python
# Đổi unix time sang datetime
date = pd.to_datetime(dataset["time"], unit="ms")
date
```

```python
# Xác định quarter của data
dataset["quarter"] = [x.quarter for x in date]
```

```python
# Tính mean theo từng quarter
print(dataset.groupby("quarter")["mag"].mean())
```
